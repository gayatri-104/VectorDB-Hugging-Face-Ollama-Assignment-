# Question 1: What is a Vector Database (VectorDB) and how is it different from traditional databases?

## Answer

A **Vector Database (VectorDB)** is a specialized database designed to store, manage, and search high-dimensional vector embeddings generated by machine learning models. These embeddings represent data such as text, images, audio, and videos in numerical form, enabling semantic similarity searches.

### Difference Between VectorDB and Traditional Databases

| Feature        | Vector Database                       | Traditional Database                      |
| -------------- | ------------------------------------- | ----------------------------------------- |
| Data Storage   | Stores vector embeddings              | Stores structured data (rows and columns) |
| Search Method  | Similarity search (nearest neighbor)  | Exact match and SQL queries               |
| Data Type      | Unstructured and semi-structured data | Structured data                           |
| AI Integration | Designed for AI/ML applications       | Limited AI capabilities                   |
| Retrieval      | Semantic search based on meaning      | Keyword-based search                      |

### Example

If a user searches for **"cars with eco-friendly engines"**, a VectorDB can retrieve documents related to **electric vehicles**, even if those exact words are not present. A traditional database typically relies on exact keyword matching.

### Conclusion

Vector Databases are essential for modern AI applications such as semantic search, recommendation systems, Retrieval-Augmented Generation (RAG), and conversational AI.

# Question 2: Explain the various types of VectorDBs available and describe their suitability for different use cases.

## Answer

Several Vector Databases are available, each designed for specific AI and machine learning requirements.

### 1. ChromaDB

**Description:**
An open-source vector database designed specifically for AI and LLM applications.

**Use Cases:**

* RAG systems
* Chatbots
* Document search
* Local AI projects

**Advantages:**

* Easy to use
* Lightweight
* Python-friendly

---

### 2. Pinecone

**Description:**
A fully managed cloud-based vector database.

**Use Cases:**

* Enterprise AI applications
* Large-scale semantic search
* Production deployments

**Advantages:**

* High scalability
* Managed infrastructure
* Real-time updates

---

### 3. Weaviate

**Description:**
An open-source vector database with built-in machine learning modules.

**Use Cases:**

* Knowledge graphs
* Semantic search
* Recommendation engines

**Advantages:**

* Hybrid search support
* GraphQL API
* Cloud and self-hosted options

---

### 4. Milvus

**Description:**
A highly scalable open-source vector database optimized for large datasets.

**Use Cases:**

* Image search
* Video retrieval
* Large-scale AI systems

**Advantages:**

* Billions of vectors support
* High performance
* Distributed architecture

---

### 5. Qdrant

**Description:**
A vector similarity search engine focused on filtering and metadata support.

**Use Cases:**

* Recommendation systems
* Personalized search
* AI assistants

**Advantages:**

* Fast retrieval
* Strong filtering capabilities
* Easy deployment

---

### Conclusion

Different VectorDBs serve different needs. ChromaDB is ideal for learning and prototyping, Pinecone for enterprise cloud deployments, Milvus for large-scale datasets, Weaviate for semantic knowledge systems, and Qdrant for recommendation-based applications.

````python
!pip install chromadb sentence-transformers
import chromadb
from sentence_transformers import SentenceTransformer

# Create ChromaDB client
client = chromadb.Client()

# Create collection
collection = client.create_collection("my_documents")

# Sample documents
documents = [
    "Artificial Intelligence is transforming industries.",
    "Machine Learning is a subset of AI.",
    "Deep Learning uses neural networks."
]

# Load embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Generate embeddings
embeddings = model.encode(documents).tolist()

# Insert data
collection.add(
    ids=["1", "2", "3"],
    documents=documents,
    embeddings=embeddings
)

# Query
query = "What is Artificial Intelligence?"
query_embedding = model.encode([query]).tolist()

results = collection.query(
    query_embeddings=query_embedding,
    n_results=2
)

print(results["documents"])
````

````python
!pip install sentence-transformers -q

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

documents = [
    "AI stands for Artificial Intelligence.",
    "Machine Learning is a subset of AI.",
    "Deep Learning uses neural networks."
]

model = SentenceTransformer('all-MiniLM-L6-v2')

doc_embeddings = model.encode(documents)

query = "What is Machine Learning?"

query_embedding = model.encode([query])

similarities = cosine_similarity(
    query_embedding,
    doc_embeddings
)

best_match = documents[similarities.argmax()]

print("Question:", query)
print("Retrieved Document:", best_match)
````

````python
!pip install sentence-transformers -q

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

articles = [
    "Diabetes is a disease characterized by high blood sugar levels.",
    "Hypertension increases the risk of heart disease.",
    "Regular exercise improves cardiovascular health."
]

model = SentenceTransformer('all-MiniLM-L6-v2')

article_embeddings = model.encode(articles)

question = "What causes high blood sugar?"

question_embedding = model.encode([question])

similarity = cosine_similarity(
    question_embedding,
    article_embeddings
)

answer = articles[similarity.argmax()]

print("Question:", question)
print("Relevant Article:", answer)
'''Medical Research Articles
          │
          ▼
Text Chunking
          │
          ▼
Hugging Face Embedding Model
(all-MiniLM-L6-v2)
          │
          ▼
ChromaDB Vector Database
          │
          ▼
Semantic Retrieval
          │
          ▼
Ollama + Llama3
          │
          ▼
Medical Chatbot Response'''
````
